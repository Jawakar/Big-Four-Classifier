```python
# Cleaning the directory
display_image_cleaner("/content/gdrive/MyDrive/BigFour/snake_types")
```

```python
fns = get_image_files('/content/gdrive/MyDrive/BigFour/snake_types')
# Check and remove file which are not image files
failed = verify_images(fns)
```

```python
# Creating datablock API
snake_types = DataBlock(
    blocks = (ImageBlock, CategoryBlock),
    get_items = get_image_files,
    splitter = RandomSplitter(valid_pct=0.3, seed=42),
    get_y = parent_label,
    item_tfms = RandomResizedCrop(224, min_scale=0.5),
    batch_tfms = aug_transforms()
)
```

```python
dls = snake_types.dataloaders('/content/gdrive/MyDrive/BigFour/snake_types')
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
# using cnn architecture with resnet18 for training
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(12)
```

```python
# confusion matrix
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
# export and save the model
learn.path = Path('/content/gdrive/MyDrive/BigFour')
learn.export("BigFour.pkl")
```
